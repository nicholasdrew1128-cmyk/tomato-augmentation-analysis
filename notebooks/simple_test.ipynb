```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("ashishmotwani/tomato")

print("Path to dataset files:", path)
```

```python
import matplotlib.pyplot as plt
from PIL import Image
import glob
import os

image_files = glob.glob(f"{path}/**/*.jpg", recursive=True)

if image_files:
    sample_path = image_files[0]
    img = Image.open(sample_path)

    plt.figure(figsize=(8, 6))
    plt.imshow(img)
    plt.title(f"Sample: {os.path.basename(sample_path)}")
    plt.axis('off')
    plt.show()
else:
    print("No images found. Check the file extensions in the dataset!")
```
